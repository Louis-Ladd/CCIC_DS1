```python
import time
import random
import threading 
from threading import Thread
from IPython.display import clear_output

numbers = []
for i in range(0,1000):
    numbers.append(random.randrange(0,4))
print(numbers)
sorted_numbers = []

class worker(threading.Thread):
    x = 0
    def __init__(self,in_x):
        Thread.__init__(self)
        self.value = 0
        self.x = in_x
    def run(self):
        time.sleep(self.x)
        sorted_numbers.append(self.x)

def sleep_sort(nums):
    threads = []
    for i in range(0,len(nums)):
        thread = worker(nums[i])
        threads.append(thread)
        thread.start()
    for thread in threads:
        thread.join()
    
sleep_sort(numbers)
clear_output(wait=True)
print(sorted_numbers)
```
